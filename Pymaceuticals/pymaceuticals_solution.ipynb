### Analysis of Drug Regimens and Tumor Volume Data

Distribution of Mice Across Drug Regimen:
- The bar graph analysis revealed that the Capomulin drug regimen had the largest number of treated mice, with 230 entries, while Propriva had the smallest, with 148 entries.
- After removing duplicates, the total number of unique mice in the dataset was 248.
- A gender analysis showed a fairly balanced distribution of mice with only a 2% difference between the genders. 
	
Correlation Between Mouse Weight and Tumor Volume:
- The correlation coefficient between mouse weight and average tumor volume was calculated as 0.84, indicating a strong positive correlation.
- This suggests that as the weight of a mouse increases, its average tumor volume also tends to increase proportionally.
	
Effectiveness of Treatments:
- Among the tested drug regimens, Capomulin and Ramicane stood out as the most effective treatments. Both regimens showed better tumor size reduction compared to other drugs.
- Mice treated with Capomulin and Ramicane had consistently lower final average tumor volumes, highlighting their efficacy.
	
Insights from Capomulin Treatment:
- Weight-Tumor Volume Relationship: For mice treated with Capomulin, a strong positive correlation between weight and tumor volume was observed, consistent with the overall dataset trend. This means heavier mice tended to have larger tumors, even under Capomulin treatment.
- Mouse l509 Case Study: A detailed examination of Mouse l509 showed that its tumor volumes decreased significantly over time while undergoing Capomulin treatment, showcasing the drugâ€™s effectiveness in reducing tumor size.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data_complete =pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
study_data_complete.head(10)
```

```python
# Checking the number of mice. - use .unique()
len(study_data_complete["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse_ids = study_data_complete.loc[study_data_complete.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_mouse_ids[0]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data = study_data_complete.loc[study_data_complete["Mouse ID"] == duplicate_mouse_ids[0]]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# use .isin()
clean_study_data_complete = study_data_complete[study_data_complete["Mouse ID"].isin(duplicate_mouse_ids)==False]
clean_study_data_complete.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_study_data_complete["Mouse ID"].unique())
```

## Summary Statistics

```python
numeric_columns = clean_study_data_complete.select_dtypes(include='number')
means = clean_study_data_complete.groupby("Drug Regimen")[numeric_columns.columns].mean()
means["Tumor Volume (mm3)"]
```

```python
numeric_columns = clean_study_data_complete.select_dtypes(include='number')
medians = clean_study_data_complete.groupby("Drug Regimen")[numeric_columns.columns].median()
medians["Tumor Volume (mm3)"]
```

```python
numeric_columns = clean_study_data_complete.select_dtypes(include='number')
variances = clean_study_data_complete.groupby("Drug Regimen")[numeric_columns.columns].var()
variances["Tumor Volume (mm3)"]
```

```python
numeric_columns = clean_study_data_complete.select_dtypes(include='number')
stds = clean_study_data_complete.groupby("Drug Regimen")[numeric_columns.columns].std()
stds["Tumor Volume (mm3)"]
```

```python
numeric_columns = clean_study_data_complete.select_dtypes(include='number')
sems = clean_study_data_complete.groupby("Drug Regimen")[numeric_columns.columns].sem()
sems["Tumor Volume (mm3)"]
```

```python
# generate a summary dictionary
summaryDict = {
    "Mean Tumor Volume": means["Tumor Volume (mm3)"],
    "Median Tumor Volume": medians["Tumor Volume (mm3)"],
    "Tumor Volume Variance": variances["Tumor Volume (mm3)"],
    "Tumor Volume Std. Dev.": stds["Tumor Volume (mm3)"],
    "Tumor Volume Std. Err.": sems["Tumor Volume (mm3)"]
}

df = pd.DataFrame(summaryDict)
df    
```

## Quartiles, Outliers and Boxplots

```python
# Start by getting the last (greatest) timeprint for each mouse (max timepoint)
# group by mouse ID -> get the max of the Timepoint column
#clean_study_data_complete.head()
maxTime = clean_study_data_complete.groupby("Mouse ID")["Timepoint"].max() # creates a series
# to covert the series to a dataframe, reset the index of the series
maxTime = maxTime.reset_index()
maxTime.head()
```

```python
clean_study_data_complete.head()
```

```python
# merge to keep data for the max  timepoints of each mouse only
merged_data = maxTime.merge(clean_study_data_complete, on=["Mouse ID", "Timepoint"], how="left")
merged_data.head()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
```

```python
for treatment in treatment_list:
    # Locate the rows which contain mice on each drug and get the tumor volumes - use .loc()
    info = merged_data.loc[merged_data["Drug Regimen"] == treatment]

info
```

```python
for treatment in treatment_list:
    # Locate the rows which contain mice on each drug and get the tumor volumes - use .loc()
    final_tumor_vol = merged_data.loc[merged_data["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]

final_tumor_vol
```

```python
# Create / recreate empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

for treatment in treatment_list:
    # Locate the rows which contain mice on each drug and get the tumor volumes - use .loc()
    final_tumor_vol = merged_data.loc[merged_data["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]

    # add each series to the empty tumor volume list
    tumor_vol_list.append(final_tumor_vol)

tumor_vol_list # prints a list of series
```

```python
# Create / recreate empty list to fill with tumor volume data (for plotting)
tumor_vol_list = []

for treatment in treatment_list:
    # Locate the rows which contain mice on each drug and get the tumor volumes - use .loc()
    final_tumor_vol = merged_data.loc[merged_data["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]

    # Add each series to the empty tumor volume list
    tumor_vol_list.append(final_tumor_vol)

    # Determine outliers using upper and lower bounds
    quartiles = final_tumor_vol.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)  # Corrected to upper_bound

    # Identify potential outliers
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    
    # Print potential outliers
    if outliers.empty:
        print(f"{treatment}'s potential outliers: None")
    else:
        outlier_list = ', '.join([str(o) for o in outliers])
        print(f"{treatment}'s potential outliers: {outlier_list}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
# Customize the outliers
flierprops = dict(marker='o', markerfacecolor='gold', markersize=16,
                  markeredgecolor='black')
# Customize the median line
medianprops = dict(linestyle='-', linewidth=2.5, color='gold')
# Customize the mean point
meanpointprops = dict(marker='D', markeredgecolor='black',
                      markerfacecolor='gold')

plt.boxplot(tumor_vol_list, tick_labels=treatment_list,
            flierprops=flierprops, medianprops=medianprops, meanprops=meanpointprops,
            showmeans=True)

plt.ylabel('Final Tumor Volume (mm3)')
plt.title("Distribution of the Tumor Volume for Each Treatment Group")
plt.show()
```

## Correlation and Regression

```python
corr = st.pearsonr(capomulinAveragesDF['Weight (g)'],capomulinAveragesDF['Tumor Volume (mm3)'])[0]
print(f"The correlation between mouse weight and the average tumor volume is {corr:.2f}")
```

```python
x_values = capomulinAveragesDF["Weight (g)"] 
y_values = capomulinAveragesDF["Tumor Volume (mm3)"] 

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values) 

regressionValues = x_values * slope + intercept 
lineEQ = f"y = {slope:.2f}x + {intercept:.2f}"
```

```python
plt.scatter(x_values, y_values, facecolor="gold", edgecolor="black", s=50)

plt.plot(x_values, regressionValues, "r-")
plt.annotate(lineEQ, (20, 37), fontsize=16, color="red")

plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs. Mouse Weight for the Capomulin Regimen")
plt.show()           
```
